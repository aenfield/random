```python
import dask.distributed
from dask import delayed
```

```python
client = dask.distributed.Client()
client
```

```python
def inc(x):
    return x + 1

def double(x):
    return x * 2

def add(x, y):
    return x + y
```

```python
data = range(10,16)
```

```python
output = []
for x in data:
    a = inc(x)
    b = double(x)
    c = add(a, b)
    output.append(c)
    
total = sum(output)
total
```

```python
output = []
for x in data:
    a = delayed(inc)(x)
    b = delayed(double)(x)
    c = delayed(add)(a, b)
    output.append(c)
    
total = delayed(sum)(output)
```

```python
total.visualize()
```

```python
total.compute()
```
