# Implement perceptron?

```python
learning_rate = 0.1
```

```python
def y_hat(w, x):
    return w.T.dot(x)
```

```python
#X = np.array([1,1,0,1,2,1,1,-1,-1]).reshape(3,3)
X = np.array([5,10,0,10,20,0,30,60,0]).reshape(3,3)

#y = np.array([0,0,0])
y = np.array([1,1,0])

w = np.zeros(9).reshape(3,3)

print(X)
print(y)
print(w)
```

```python
for i in range(3):
    yhi = y_hat(w[i], X[i])
    for j in range(3):
        w[i,j] = w[i,j] + learning_rate*(yhi - y[i])*X[i,j]
    print("Updated weights for output neuron {}".format(i))
    print(w)
```

```python
print([y_hat(w[i], X[i]) for i in range(3)])
print(y)
```
